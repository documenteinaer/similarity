```python
import csv
import os
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# with open('acc-gyr-mag-2022-01-13.csv', newline='') as csvfile:
#     fin = csv.reader(csvfile, delimiter=',', quotechar='"')
#     for row in fin:
#         print('='.join(row))

# time,gFx,gFy,gFz,wx,wy,wz,Bx,By,Bz


datadir = "../data.upb/magnetometer/"
datafile = 'acc-gyr-mag-2022-01-13.csv'
md = np.genfromtxt(os.path.join(datadir, datafile), delimiter=',')
md = np.delete(md, 0, 0)
nr = md.shape[0]
print(md[nr-1, :])
```

```python

plt.style.use("seaborn-whitegrid")
fig, ax = plt.subplots(figsize=(12, 8))
plt.title(f"Rotating chair, phone fixed in vertical position")    
plt.legend(loc="lower left",shadow=True, fancybox=True)


plt.subplot(2, 2, 1)
plt.xlabel('Time[s]')
plt.ylabel('accelerometer $m/s^2$')
plt.grid(True)
plt.plot(md[:, 0], md[:, 1], "-", label='gX')
plt.plot(md[:, 0], md[:, 2], "-", label='gY')
plt.plot(md[:, 0], md[:, 3], "-", label='gZ')
plt.legend(loc="lower left",shadow=True, fancybox=True)

plt.subplot(2, 2, 2)
plt.xlabel('Time[s]')
plt.ylabel('rotation speed rad/s')
plt.grid(True)
plt.legend(loc="lower left",shadow=True, fancybox=True)
plt.plot(md[:, 0], md[:, 4], "-", label='wx')
plt.plot(md[:, 0], md[:, 5], "-", label='wy')
plt.plot(md[:, 0], md[:, 6], "-", label='wz')
plt.legend(loc="lower left",shadow=True, fancybox=True)

plt.subplot(2, 2, 3)
plt.xlabel('Time[s]')
plt.ylabel('Magnetic field $\mu T$')
plt.grid(True)
plt.plot(md[:, 0], md[:, 7], "-", label='Bx')
plt.plot(md[:, 0], md[:, 8], "-", label='By')
plt.plot(md[:, 0], md[:, 9], "-", label='Bz')
plt.legend(loc="lower left",shadow=True, fancybox=True)

#plt.subplot(2, 2, 4)
#plt.plot(md[:, 0], md[:, 9], "-", label='Bz')  

#plt.xticks(range(0, 10))
# plt.yticks(np.arange(-0.1, 1, 0.1))
#ax.set_ylim(0, 0.7)
#ax.set_xlim(-0.1, 10.1)
plt.legend(loc="lower left",shadow=True, fancybox=True)
plt.show()
# fig.savefig(f"Braycurtis_vs_dist_precis_rnd2dir.pdf", bbox_inches='tight')
```
