```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('02-precis.xyz',sep='\s+',header=None)
data = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(6, 6))
img = plt.imread("arch_precis_etaj2.png")
ax.imshow(img, alpha=0.5, extent=[36.5, 59.5, 38.5, 15])
plt.xlabel('[Meters]')
plt.ylabel('[Meters]')
plt.plot(data[2], data[3],'bo', marker='s', markersize=3)
plt.show()
fig.savefig(f"precis-02-topology.pdf", bbox_inches='tight')
```
